在MuJoCo中，有几种主要类型的执行器（actuator）。以下是MuJoCo中可用的主要执行器类型：

1. 通用力（general force）：
   - 标签：<general>
   - 应用指定的力或力矩到关节或自由度。

2. 电机（motor）：
   - 标签：<motor>
   - 模拟电机行为，可以应用力或力矩。

3. 位置伺服（position servo）：
   - 标签：<position>
   - 控制关节位置。

4. 速度伺服（velocity servo）：
   - 标签：<velocity>
   - 控制关节速度。

5. 肌肉（muscle）：
   - 标签：<muscle>
   - 模拟生物肌肉的行为。

6. 粘性（cylinder）：
   - 标签：<cylinder>
   - 模拟液压或气压缸。

7. 粘性（adhesion）：
   - 标签：<adhesion>
   - 模拟粘性效果。

8. 无身体约束（body-wise constraint）：
   - 标签：<intvelocity>
   - 控制整个刚体的速度。

9. 接触力（contact force）：
   - 标签：<contact>
   - 在指定的接触点施加力。

每种执行器类型都有其特定的属性和用途。选择合适的执行器类型取决于你要模拟的系统和控制策略。

在MuJoCo的XML模型文件中，这些执行器通常定义在<actuator>部分。例如：

```xml
<mujoco>
  <actuator>
    <motor joint="joint1" name="motor1" gear="100"/>
    <position joint="joint2" name="servo1" kp="10"/>
    <velocity joint="joint3" name="vel_ctrl1" kv="5"/>
  </actuator>
</mujoco>
```

这个例子定义了一个电机、一个位置伺服和一个速度伺服执行器。

需要注意的是，MuJoCo的版本可能会影响可用的执行器类型和具体属性，因此在使用时最好参考与你所使用版本相对应的官方文档。
